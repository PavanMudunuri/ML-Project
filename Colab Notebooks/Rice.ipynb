```python
# ============================
# STEP 0: Mount Drive & Imports
# ============================
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0, MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Concatenate, Dropout, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.metrics import Precision, Recall

# ============================
# STEP 1: Dataset Paths
# ============================
dataset_dir = "/content/drive/MyDrive/CapStoneProjectDataSets/Rice_split"

# Temporary directories for train/valid split
base_dir = "/content/rice_split"
train_dir = os.path.join(base_dir, "train")
valid_dir = os.path.join(base_dir, "valid")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)

if not os.path.exists(dataset_dir):
    raise FileNotFoundError(f"Dataset not found at {dataset_dir}")

# ============================
# STEP 2: Split dataset (80/20)
# ============================
split_ratio = 0.8

for disease in os.listdir(dataset_dir):
    disease_path = os.path.join(dataset_dir, disease)
    if not os.path.isdir(disease_path):
        continue

    os.makedirs(os.path.join(train_dir, disease), exist_ok=True)
    os.makedirs(os.path.join(valid_dir, disease), exist_ok=True)

    images = [f for f in os.listdir(disease_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    random.shuffle(images)

    split_idx = int(len(images) * split_ratio)
    train_images = images[:split_idx]
    valid_images = images[split_idx:]

    for img in train_images:
        shutil.copy(os.path.join(disease_path, img), os.path.join(train_dir, disease, img))
    for img in valid_images:
        shutil.copy(os.path.join(disease_path, img), os.path.join(valid_dir, disease, img))

print("✅ Dataset split complete")

# ============================
# STEP 3: Data Generators
# ============================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

valid_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True,
    seed=123
)

valid_gen = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

class_names = list(train_gen.class_indices.keys())
print("Classes:", class_names)

# ============================
# STEP 4: Build Hybrid Model
# ============================
input_tensor = Input(shape=(224, 224, 3))

effnet = EfficientNetB0(weights="imagenet", include_top=False, input_tensor=input_tensor)
mobilenet = MobileNetV2(weights="imagenet", include_top=False, input_tensor=input_tensor)

# Freeze base models
effnet.trainable = False
mobilenet.trainable = False

effnet_out = GlobalAveragePooling2D()(effnet.output)
mobilenet_out = GlobalAveragePooling2D()(mobilenet.output)

merged = Concatenate()([effnet_out, mobilenet_out])
x = Dense(512, activation="relu")(merged)
x = Dropout(0.5)(x)
output = Dense(len(class_names), activation="softmax")(x)

model = Model(inputs=input_tensor, outputs=output)

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy", Precision(name="precision"), Recall(name="recall")]
)

model.summary()

# ============================
# STEP 5: Training
# ============================
EPOCHS = 100
callbacks = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=5, min_lr=1e-6),
    ModelCheckpoint("best_hybrid_rice.h5", save_best_only=True, monitor="val_loss")
]

history = model.fit(
    train_gen,
    validation_data=valid_gen,
    epochs=EPOCHS,
    callbacks=callbacks
)

# ============================
# STEP 6: Plot Training Curves
# ============================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend(); plt.title("Accuracy")

plt.subplot(1,2,2)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend(); plt.title("Loss")
plt.show()

# ============================
# STEP 7: Evaluate Model
# ============================
val_loss, val_acc, val_prec, val_rec = model.evaluate(valid_gen, verbose=0)

train_acc = history.history["accuracy"][-1]
train_prec = history.history["precision"][-1]
train_rec = history.history["recall"][-1]
train_loss = history.history["loss"][-1]

print("\n🔹 Training Performance:")
print(f"  Accuracy:  {train_acc:.4f}")
print(f"  Precision: {train_prec:.4f}")
print(f"  Recall:    {train_rec:.4f}")
print(f"  Loss:      {train_loss:.4f}")

print("\n🔹 Validation Performance:")
print(f"  Accuracy:  {val_acc:.4f}")
print(f"  Precision: {val_prec:.4f}")
print(f"  Recall:    {val_rec:.4f}")
print(f"  Loss:      {val_loss:.4f}")

model.save("rice_hybrid_final.h5")
print("\n✅ Model saved as 'rice_hybrid_final.h5'")
```

```python
# ==========================
# STEP 0: Imports
# ==========================
import os
import random
import shutil
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import load_model
from tensorflow.keras.metrics import Precision, Recall

# ==========================
# STEP 1: Paths and Model
# ==========================
MODEL_PATH = "/content/drive/MyDrive/rice_hybrid_final.h5"    # ✅ trained model
DATASET_DIR = "/content/drive/MyDrive/CapStoneProjectDataSets/Rice"   # ✅ full dataset
TEMP_DIR = "/content/rice_true_eval_split"                            # fresh temp split folder

FINE_TUNE_DIR = os.path.join(TEMP_DIR, "fine_tune")
FINAL_TEST_DIR = os.path.join(TEMP_DIR, "final_test")

# clear any old data
if os.path.exists(TEMP_DIR):
    shutil.rmtree(TEMP_DIR)

os.makedirs(FINE_TUNE_DIR, exist_ok=True)
os.makedirs(FINAL_TEST_DIR, exist_ok=True)

print(f"✅ Using model: {MODEL_PATH}")
model = load_model(MODEL_PATH)

# recompile to restore metrics
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy", Precision(name="precision"), Recall(name="recall")])

# ==========================
# STEP 2: Create 10% + 10% Split (No overlap)
# ==========================
for cls in os.listdir(DATASET_DIR):
    cls_path = os.path.join(DATASET_DIR, cls)
    if not os.path.isdir(cls_path):
        continue

    images = [f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if len(images) < 10:
        continue

    random.shuffle(images)
    total = len(images)

    fine_tune_count = int(total * 0.1)
    test_count = int(total * 0.1)

    fine_tune_imgs = images[:fine_tune_count]
    test_imgs = images[fine_tune_count:fine_tune_count + test_count]

    fine_cls_dir = os.path.join(FINE_TUNE_DIR, cls)
    test_cls_dir = os.path.join(FINAL_TEST_DIR, cls)
    os.makedirs(fine_cls_dir, exist_ok=True)
    os.makedirs(test_cls_dir, exist_ok=True)

    # ✅ copy (not symlink) to ensure independence
    for img in fine_tune_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(fine_cls_dir, img))
    for img in test_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(test_cls_dir, img))

print("✅ Created real 10% fine-tune and 10% test subsets (no overlap).")

# ==========================
# STEP 3: Data Generators
# ==========================
IMG_SIZE = (224, 224)
BATCH_SIZE = 16

datagen = ImageDataGenerator(rescale=1. / 255)

fine_tune_gen = datagen.flow_from_directory(
    FINE_TUNE_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical"
)

test_gen = datagen.flow_from_directory(
    FINAL_TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

class_names = list(test_gen.class_indices.keys())
print("✅ Classes detected:", class_names)

# ==========================
# STEP 4: Evaluate on True Test Set
# ==========================
print("\n📊 Evaluating model on unseen 10% test subset...")
val_loss, val_acc, val_prec, val_rec = model.evaluate(test_gen, verbose=1)

print("\n📈 Evaluation Results:")
print(f"✅ Test Accuracy:  {val_acc:.4f}")
print(f"✅ Test Precision: {val_prec:.4f}")
print(f"✅ Test Recall:    {val_rec:.4f}")
print(f"✅ Test Loss:      {val_loss:.4f}")

# ==========================
# STEP 5: Random Predictions
# ==========================
print("\n🔄 Showing predictions for 2 random test images...")

for i in range(2):
    rand_class = random.choice(class_names)
    rand_folder = os.path.join(FINAL_TEST_DIR, rand_class)
    valid_images = [f for f in os.listdir(rand_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if not valid_images:
        continue

    rand_image = random.choice(valid_images)
    img_path = os.path.join(rand_folder, rand_image)

    img = load_img(img_path, target_size=IMG_SIZE)
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    preds = model.predict(img_array, verbose=0)
    pred_idx = np.argmax(preds[0])
    pred_class = class_names[pred_idx]
    confidence = np.max(preds[0])

    if "healthy" in pred_class.lower():
        status = "✅ Healthy"
    else:
        status = f"⚠️ Unhealthy ({pred_class})"

    print(f"\n--- Prediction {i + 1} ---")
    print(f"🖼 Image: {rand_image}")
    print(f"Prediction: {status}")
    print(f"📊 Confidence: {confidence:.4f}")

    plt.imshow(img)
    plt.title(f"{status} ({confidence:.2f})")
    plt.axis("off")
    plt.show()
```

```python
# ======================================================
# Use Rice .h5 model (preferred) -> create final_test -> print only specificities
# ======================================================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, glob, random, shutil
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix

# -----------------------
# Config (edit if needed)
# -----------------------
DRIVE_ROOT = "/content/drive/MyDrive"
PREFERRED_H5_NAME = "rice_hybrid.h5"   # preferred exact filename (change if yours differs)
FALLBACK_KEYWORD = "rice"              # will match any .h5 containing this keyword if exact not found

SRC_DATASET = "/content/drive/MyDrive/CapStoneProjectDataSets/Rice"  # rice dataset location (per-class folders)
TEMP_ROOT = "/content/rice_true_eval_split"
FINAL_TEST_DIR = os.path.join(TEMP_ROOT, "final_test")
IMG_SIZE = (224, 224)
BATCH_SIZE = 16
TEST_FRAC = 0.10
SEED = 42
IMG_EXTS = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp')

random.seed(SEED)

# -----------------------
# Find candidate .h5 files
# -----------------------
h5_paths = glob.glob(os.path.join(DRIVE_ROOT, "**", "*.h5"), recursive=True)
print(f"Found {len(h5_paths)} .h5 files under {DRIVE_ROOT}.")

# Prefer exact filename if present
MODEL_PATH = None
for p in h5_paths:
    if os.path.basename(p) == PREFERRED_H5_NAME:
        MODEL_PATH = p
        break

# If not found, prefer any .h5 that contains FALLBACK_KEYWORD
if MODEL_PATH is None:
    for p in h5_paths:
        if FALLBACK_KEYWORD.lower() in os.path.basename(p).lower():
            MODEL_PATH = p
            break

# If still none, raise error
if MODEL_PATH is None:
    raise FileNotFoundError("No Rice .h5 model found under Drive. Place your .h5 in Drive or update FALLBACK_KEYWORD.")

print("Selected model:", MODEL_PATH)

# -----------------------
# Ensure final_test exists (create fresh 10% per-class if needed)
# -----------------------
if os.path.exists(TEMP_ROOT):
    shutil.rmtree(TEMP_ROOT)
os.makedirs(FINAL_TEST_DIR, exist_ok=True)

if not os.path.exists(SRC_DATASET):
    raise FileNotFoundError(f"Rice dataset not found at {SRC_DATASET}")

# Copy 10% per class (min 1) into FINAL_TEST_DIR
for cls in sorted(os.listdir(SRC_DATASET)):
    cls_src = os.path.join(SRC_DATASET, cls)
    if not os.path.isdir(cls_src):
        continue
    imgs = [f for f in os.listdir(cls_src) if f.lower().endswith(IMG_EXTS)]
    if not imgs:
        continue
    random.shuffle(imgs)
    n = max(1, int(len(imgs) * TEST_FRAC))
    selected = imgs[:n]
    dst_cls = os.path.join(FINAL_TEST_DIR, cls)
    os.makedirs(dst_cls, exist_ok=True)
    for fn in selected:
        shutil.copy2(os.path.join(cls_src, fn), os.path.join(dst_cls, fn))

# quick sanity
total_test_images = sum(len(files) for _, _, files in os.walk(FINAL_TEST_DIR))
if total_test_images == 0:
    raise RuntimeError(f"No test images created under {FINAL_TEST_DIR}. Check {SRC_DATASET} structure.")
print(f"Created test folder: {FINAL_TEST_DIR} with {total_test_images} images.")

# -----------------------
# Load model (Rice .h5)
# -----------------------
model = load_model(MODEL_PATH)
print("Model loaded.")

# -----------------------
# Create generator & predict
# -----------------------
datagen = ImageDataGenerator(rescale=1./255)
test_gen = datagen.flow_from_directory(
    FINAL_TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
if test_gen.samples == 0:
    raise RuntimeError("No images found by generator. Check FINAL_TEST_DIR structure.")

preds = model.predict(test_gen, verbose=0)
y_pred = np.argmax(preds, axis=1)
y_true = test_gen.classes
class_names = list(test_gen.class_indices.keys())

# -----------------------
# Compute specificity only
# -----------------------
cm = confusion_matrix(y_true, y_pred)
eps = 1e-7
specificities = []
total = cm.sum()
for i in range(len(cm)):
    TP = cm[i, i]
    FP = cm[:, i].sum() - TP
    FN = cm[i, :].sum() - TP
    TN = total - TP - FP - FN
    spec = TN / (TN + FP + eps)
    specificities.append(spec)
# Print only specificities

print(f"Specificity: {np.mean(specificities)*100:.2f}%")
```

```python
# ==========================
# STEP 0: Imports
# ==========================
import os
import random
import shutil
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import load_model
from tensorflow.keras.metrics import Precision, Recall

# ==========================
# STEP 1: Paths and Model
# ==========================
MODEL_PATH = "/content/drive/MyDrive/rice_hybrid.h5"    # ✅ trained model
DATASET_DIR = "/content/drive/MyDrive/CapStoneProjectDataSets/Rice"   # ✅ full dataset
TEMP_DIR = "/content/rice_true_eval_split"                            # fresh temp split folder

FINE_TUNE_DIR = os.path.join(TEMP_DIR, "fine_tune")
FINAL_TEST_DIR = os.path.join(TEMP_DIR, "final_test")

# clear any old data
if os.path.exists(TEMP_DIR):
    shutil.rmtree(TEMP_DIR)

os.makedirs(FINE_TUNE_DIR, exist_ok=True)
os.makedirs(FINAL_TEST_DIR, exist_ok=True)

print(f"✅ Using model: {MODEL_PATH}")
model = load_model(MODEL_PATH)

# recompile to restore metrics
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy", Precision(name="precision"), Recall(name="recall")])

# ==========================
# STEP 2: Take 10 random images per class (no overlap)
# ==========================
CLASSES = ['Bacterial leaf blight', 'Brown spot', 'Leaf smut']
IMAGES_PER_CLASS = 10  # ✅ total 30 images (10 * 3 classes)

for cls in CLASSES:
    cls_path = os.path.join(DATASET_DIR, cls)
    if not os.path.isdir(cls_path):
        continue

    images = [f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if len(images) < IMAGES_PER_CLASS:
        print(f"⚠️ Warning: {cls} has less than {IMAGES_PER_CLASS} images available.")
        continue

    random.shuffle(images)
    selected_imgs = images[:IMAGES_PER_CLASS]

    test_cls_dir = os.path.join(FINAL_TEST_DIR, cls)
    os.makedirs(test_cls_dir, exist_ok=True)

    # ✅ copy (not symlink)
    for img in selected_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(test_cls_dir, img))

print(f"✅ Created evaluation set: {IMAGES_PER_CLASS} images per class → total {IMAGES_PER_CLASS * len(CLASSES)} images.")

# ==========================
# STEP 3: Data Generators
# ==========================
IMG_SIZE = (224, 224)
BATCH_SIZE = 8

datagen = ImageDataGenerator(rescale=1. / 255)

test_gen = datagen.flow_from_directory(
    FINAL_TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

class_names = list(test_gen.class_indices.keys())
print("✅ Classes detected:", class_names)

# ==========================
# STEP 4: Evaluate on True Test Set
# ==========================
print("\n📊 Evaluating model on 30-image (10 per class) test subset...")
val_loss, val_acc, val_prec, val_rec = model.evaluate(test_gen, verbose=1)

print("\n📈 Evaluation Results:")
print(f"✅ Test Accuracy:  {val_acc:.4f}")
print(f"✅ Test Precision: {val_prec:.4f}")
print(f"✅ Test Recall:    {val_rec:.4f}")
print(f"✅ Test Loss:      {val_loss:.4f}")

# ==========================
# STEP 5: Predictions on All 30 Test Images
# ==========================
print("\n🔄 Showing predictions for all 30 test images...\n")

for cls in CLASSES:
    print(f"\n===== Predictions for Class: {cls} =====")
    cls_folder = os.path.join(FINAL_TEST_DIR, cls)
    valid_images = [f for f in os.listdir(cls_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    for img_name in valid_images:
        img_path = os.path.join(cls_folder, img_name)
        img = load_img(img_path, target_size=IMG_SIZE)
        img_array = img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        preds = model.predict(img_array, verbose=0)
        pred_idx = np.argmax(preds[0])
        pred_class = class_names[pred_idx]
        confidence = np.max(preds[0])

        if "healthy" in pred_class.lower():
            status = "✅ Healthy"
        else:
            status = f"⚠️ Unhealthy ({pred_class})"

        print(f"🖼 {img_name} → {status} | Confidence: {confidence:.4f}")

        plt.imshow(img)
        plt.title(f"{pred_class} ({confidence:.2f})")
        plt.axis("off")
        plt.show()
```
