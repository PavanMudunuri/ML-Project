```python
# ============================
# STEP 0: Mount Drive & Imports
# ============================
from google.colab import drive
drive.mount('/content/drive')

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0, MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Concatenate, Dropout, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.metrics import Precision, Recall

# âœ… Enable mixed precision for faster GPU training
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# ============================
# STEP 1: Dataset Path
# ============================
base_dataset_dir = "/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/taiwan"
data_type = "data_augmentation"   # âœ… Only use data_augmentation dataset

# ============================
# STEP 2: Parameters
# ============================
IMG_SIZE = (160, 160)
BATCH_SIZE = 32
EPOCHS = 100

# ============================
# STEP 3: Define and Train Model
# ============================
def train_tomato_model(data_type):
    print("\n" + "="*70)
    print(f"ğŸ… TRAINING MODEL FOR: {data_type.upper()} DATASET")
    print("="*70)

    train_dir = os.path.join(base_dataset_dir, data_type, "Train")
    valid_dir = os.path.join(base_dataset_dir, data_type, "Test")

    if not os.path.exists(train_dir):
        raise FileNotFoundError(f"âŒ Training folder not found: {train_dir}")
    if not os.path.exists(valid_dir):
        raise FileNotFoundError(f"âŒ Validation folder not found: {valid_dir}")

    print(f"âœ… Using {data_type} dataset")
    print(f"   â”£ TRAIN PATH: {train_dir}")
    print(f"   â”— VALID PATH: {valid_dir}")

    # ============================
    # STEP 4: Data Generators
    # ============================
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=25,
        width_shift_range=0.15,
        height_shift_range=0.15,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True
    )

    valid_datagen = ImageDataGenerator(rescale=1./255)

    train_gen = train_datagen.flow_from_directory(
        train_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        shuffle=True,
        seed=42
    )

    valid_gen = valid_datagen.flow_from_directory(
        valid_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        shuffle=False
    )

    class_names = list(train_gen.class_indices.keys())
    print("\nğŸ§© Classes Detected:", class_names)

    # ============================
    # STEP 5: Build Hybrid Model
    # ============================
    input_tensor = Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

    effnet = EfficientNetB0(weights="imagenet", include_top=False, input_tensor=input_tensor)
    mobilenet = MobileNetV2(weights="imagenet", include_top=False, input_tensor=input_tensor)

    # Freeze most layers for faster training
    for layer in effnet.layers[:-20]:
        layer.trainable = False
    for layer in mobilenet.layers[:-20]:
        layer.trainable = False

    effnet_out = GlobalAveragePooling2D()(effnet.output)
    mobilenet_out = GlobalAveragePooling2D()(mobilenet.output)

    merged = Concatenate()([effnet_out, mobilenet_out])
    x = Dense(512, activation="relu")(merged)
    x = Dropout(0.4)(x)
    output = Dense(len(class_names), activation="softmax", dtype='float32')(x)

    model = Model(inputs=input_tensor, outputs=output)

    model.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss="categorical_crossentropy",
        metrics=["accuracy", Precision(name="precision"), Recall(name="recall")]
    )

    print("\nğŸ§  Model Summary:")
    model.summary()

    # ============================
    # STEP 6: Training
    # ============================
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor="val_loss", factor=0.3, patience=5, min_lr=1e-6),
        ModelCheckpoint(f"best_hybrid_tomato_{data_type}.h5", save_best_only=True, monitor="val_accuracy", mode='max')
    ]

    history = model.fit(
        train_gen,
        validation_data=valid_gen,
        steps_per_epoch=20,  # âš¡ Reduced from 50 to 20 for faster epochs
        validation_steps=valid_gen.samples // BATCH_SIZE,
        epochs=EPOCHS,
        callbacks=callbacks,
        verbose=1
    )

    # ============================
    # STEP 7: Always Print Performance
    # ============================
    try:
        train_acc = history.history["accuracy"][-1]
        train_prec = history.history["precision"][-1]
        train_rec = history.history["recall"][-1]
        train_loss = history.history["loss"][-1]
    except Exception:
        train_acc = train_prec = train_rec = train_loss = 0.0

    val_loss, val_acc, val_prec, val_rec = model.evaluate(valid_gen, verbose=0)

    print("\nğŸ“Š TRAINING PERFORMANCE:")
    print(f"  Accuracy:  {train_acc:.4f}")
    print(f"  Precision: {train_prec:.4f}")
    print(f"  Recall:    {train_rec:.4f}")
    print(f"  Loss:      {train_loss:.4f}")

    print("\nğŸ“ˆ VALIDATION PERFORMANCE:")
    print(f"  Accuracy:  {val_acc:.4f}")
    print(f"  Precision: {val_prec:.4f}")
    print(f"  Recall:    {val_rec:.4f}")
    print(f"  Loss:      {val_loss:.4f}")

    # ============================
    # STEP 8: Plot Training Curves
    # ============================
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(history.history["accuracy"], label="Train Acc")
    plt.plot(history.history["val_accuracy"], label="Val Acc")
    plt.legend(); plt.title(f"Accuracy ({data_type})")

    plt.subplot(1,2,2)
    plt.plot(history.history["loss"], label="Train Loss")
    plt.plot(history.history["val_loss"], label="Val Loss")
    plt.legend(); plt.title(f"Loss ({data_type})")
    plt.show()

    # ============================
    # STEP 9: Save Model
    # ============================
    model_save_path = f"/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/tomato_hybrid_final_{data_type}.h5"
    model.save(model_save_path)
    print(f"\nâœ… Model saved successfully as: {model_save_path}")
    print("="*70 + "\n")


# ============================
# STEP 10: Train Only Once
# ============================
train_tomato_model("data_augmentation")
```

```python
# ==========================================
# STEP 0: Mount Drive & Imports
# ==========================================
from google.colab import drive
drive.mount('/content/drive')

import os
import random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array
from sklearn.model_selection import train_test_split

print("âœ… Libraries imported successfully!")
print("TensorFlow version:", tf.__version__)

# ==========================================
# STEP 1: Paths and Parameters
# ==========================================
BASE_DATA_DIR = "/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/taiwan"
DATASET_TYPES = ["data_augmentation"]
IMG_SIZE = (160, 160)
BATCH_SIZE = 32
SEED = 42

# Ensure reproducibility
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# ==========================================
# STEP 2: Evaluation Function
# ==========================================
def evaluate_tomato_dataset(data_type):
    print("\n" + "="*80)
    print(f"ğŸ… EVALUATING MODEL FOR: {data_type.upper()} DATASET")
    print("="*80)

    # Model path (.h5 format)
    model_path = f"/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/tomato_hybrid_final_{data_type}.h5"
    base_valid_dir = os.path.join(BASE_DATA_DIR, data_type, "Test")

    # Temporary directories for fine-tune/test split
    temp_dir = f"/content/tomato_{data_type}_10_10_split"
    fine_tune_dir = os.path.join(temp_dir, "fine_tune")
    final_test_dir = os.path.join(temp_dir, "final_test")

    os.makedirs(fine_tune_dir, exist_ok=True)
    os.makedirs(final_test_dir, exist_ok=True)

    # ==========================================
    # STEP 3: Load Model
    # ==========================================
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"âŒ Model not found at: {model_path}")
    print(f"âœ… Loading model: {model_path}")
    model = load_model(model_path)

    # ==========================================
    # STEP 4: Split Validation into Fine-tune (10%) + Test (10%)
    # ==========================================
    print("ğŸ”„ Splitting validation data into fine-tune (10%) and test (10%) subsets...")
    split_ratio = 0.5

    for cls in os.listdir(base_valid_dir):
        cls_path = os.path.join(base_valid_dir, cls)
        if not os.path.isdir(cls_path):
            continue

        images = [f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if len(images) < 2:
            continue

        train_imgs, test_imgs = train_test_split(images, test_size=split_ratio, random_state=SEED)

        os.makedirs(os.path.join(fine_tune_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(final_test_dir, cls), exist_ok=True)

        for img_name in train_imgs:
            src = os.path.join(cls_path, img_name)
            dst = os.path.join(fine_tune_dir, cls, img_name)
            if not os.path.exists(dst):
                os.symlink(src, dst)

        for img_name in test_imgs:
            src = os.path.join(cls_path, img_name)
            dst = os.path.join(final_test_dir, cls, img_name)
            if not os.path.exists(dst):
                os.symlink(src, dst)

    print("âœ… Fine-tune (10%) and test (10%) subsets created successfully.")

    # ==========================================
    # STEP 5: Data Generators
    # ==========================================
    datagen = ImageDataGenerator(rescale=1.0 / 255.0)

    test_gen = datagen.flow_from_directory(
        final_test_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        shuffle=False
    )

    class_names = list(test_gen.class_indices.keys())
    print(f"ğŸ§© Classes detected: {class_names}")

    # ==========================================
    # STEP 6: Evaluate Model
    # ==========================================
    print("\nğŸ“Š Evaluating model on final 10% test subset...")
    results = model.evaluate(test_gen, verbose=1)
    metric_names = model.metrics_names

    print("\nğŸ“ˆ Evaluation Results:")
    for name, value in zip(metric_names, results):
        print(f"âœ… {name.capitalize()}: {value:.4f}")

    # ==========================================
    # STEP 7: Show 10â€“20 Predictions (Different Classes)
    # ==========================================
    print("\nğŸ” Showing predictions for 10â€“20 random test images (different classes)...")

    shown_count = 0
    selected_images = []

    for cls in class_names:
        cls_folder = os.path.join(final_test_dir, cls)
        valid_images = [f for f in os.listdir(cls_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not valid_images:
            continue

        rand_image = random.choice(valid_images)
        selected_images.append((cls, os.path.join(cls_folder, rand_image)))

        if len(selected_images) >= 20:
            break

    # Shuffle final list for random order
    random.shuffle(selected_images)

    # Show predictions
    for (true_cls, img_path) in selected_images:
        img_name = os.path.basename(img_path)

        # Preprocess
        img = load_img(img_path, target_size=IMG_SIZE)
        img_array = img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        preds = model.predict(img_array, verbose=0)
        pred_idx = np.argmax(preds[0])
        pred_class = class_names[pred_idx]
        confidence = np.max(preds[0])

        print(f"\nğŸ–¼ Image: {img_name}")
        print(f"ğŸ¯ True Label: {true_cls}")
        print(f"ğŸ¤– Predicted: âš ï¸ {pred_class} ({confidence:.4f})")

        plt.figure(figsize=(7.2, 6.06))
        plt.imshow(load_img(img_path))
        plt.title(f"âš ï¸ {pred_class} ({confidence:.2f})", fontsize=12, weight='bold')
        plt.axis("off")
        plt.show()

    print("\nâœ… Displayed", len(selected_images), "predictions successfully.")

# ==========================================
# STEP 8: Run Evaluation
# ==========================================
for dtype in DATASET_TYPES:
    evaluate_tomato_dataset(dtype)
```

```python
# ----------------------------
# Specificity only â€” corrected input size
# ----------------------------
from google.colab import drive
drive.mount('/content/drive', force_remount=False)

import os
import math
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix

# ---------- Settings ----------
MODEL_PATH = "/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/taiwan/tomato_best_model_fixed.keras"
TEST_DIR   = "/content/drive/MyDrive/CapStoneProjectDataSets/Tomato/taiwan/data_augmentation/Test"
IMG_SIZE   = (224, 224)   # <<< must match model expected input
BATCH_SIZE = 16

# ---------- Sanity checks ----------
if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"Model not found: {MODEL_PATH}")
if not os.path.exists(TEST_DIR):
    raise FileNotFoundError(f"Test directory not found: {TEST_DIR}")

# ---------- Load model ----------
model = tf.keras.models.load_model(MODEL_PATH)
print(f"Loaded model: {MODEL_PATH}")

# ---------- Prepare test generator ----------
datagen = ImageDataGenerator(rescale=1.0/255.0)
test_gen = datagen.flow_from_directory(
    TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

class_names = list(test_gen.class_indices.keys())
print("Classes:", class_names)
n_samples = test_gen.samples
steps = math.ceil(n_samples / BATCH_SIZE)

# ---------- Predict ----------
preds_proba = model.predict(test_gen, steps=steps, verbose=1)
y_pred = np.argmax(preds_proba, axis=1)
y_true = test_gen.classes

# ---------- Confusion matrix ----------
cm = confusion_matrix(y_true, y_pred)
print("\nConfusion matrix (rows=true, cols=pred):\n", cm)

# ---------- Specificity per class ----------
spec_list = []
total = np.sum(cm)
for i in range(cm.shape[0]):
    TP = cm[i, i]
    FP = np.sum(cm[:, i]) - TP
    FN = np.sum(cm[i, :]) - TP
    TN = total - TP - FP - FN
    specificity = TN / (TN + FP + 1e-9)
    spec_list.append(specificity)

macro_spec = float(np.mean(spec_list))

# ---------- Print specificity only ----------
print("\nğŸ“Š Specificity (per class):")
for i, cls in enumerate(class_names):
    print(f"  {cls:20s} : {spec_list[i]*100:.2f}%")
print(f"\nâœ…  Specificity: {macro_spec*100:.2f}%")
```
